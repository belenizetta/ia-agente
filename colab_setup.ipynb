```python
# 5) Instalar dependencias
!pip install -q fastapi uvicorn python-dotenv pydantic openai gitpython PyGithub requests pyngrok
print('Dependencias instaladas.')
```

```python
# 6) Configurar variables de entorno
import os
os.environ['GITHUB_TOKEN']  = 'TU_GITHUB_TOKEN'
os.environ['GITLAB_TOKEN']  = 'TU_GITLAB_TOKEN'
os.environ['OLLAMA_BASE_URL'] = 'http://localhost:11434/v1'
os.environ['LLM_MODEL']     = 'qwen2.5-coder:7b'
os.environ['CI_COMMANDS']   = 'echo skip'
os.environ['MAX_FILES_PER_TASK'] = '5'
print('Variables configuradas.')
```

```python
# 7) Iniciar FastAPI en background
import threading
def run():
    subprocess.run(['uvicorn', 'app.main:app', '--host', '0.0.0.0', '--port', '8000'],
                   cwd='/content/ia')
threading.Thread(target=run, daemon=True).start()
import time; time.sleep(3)
print('FastAPI corriendo en puerto 8000.')
```

```python
# 8) Exponer con ngrok (obtené token gratis en ngrok.com)
from pyngrok import ngrok, conf
conf.get_default().auth_token = 'TU_NGROK_TOKEN'  # ngrok.com → tu authtoken gratis
url = ngrok.connect(8000)
print(f'\n✅ API disponible en: {url}')
print(f'\nDesde tu máquina local ejecutá:')
print(f'  python cli.py --server {url}')
```
